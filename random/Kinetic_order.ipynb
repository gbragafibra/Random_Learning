```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(Δ, S, v):
    """
    g -> kinetic order
    Δ -> derivative (dv/ds|op)
    S -> substrate concentration
    v -> rate
    """
    return Δ(Vmax, Km, S, n) * (S/v(Vmax, Km, S, n))

def v(Vmax, Km, S, n):
    #n -> Hill exponent
    return (Vmax * S**n)/(Km**n + S**n)

def Δ(Vmax, Km, S, n):
    num = (Vmax * n * S**(n-1))*(Km**n + S**n) - Vmax * S**(2*n - 1) * n
    den = (Km**n + S**n)**2
    return num/den
```

```python
Vmax = 1.5
Km = 0.2
S = np.geomspace(1e-6, 1, 50)
for n in np.linspace(1/4, 4, 10):
    plt.plot(S, g(Δ, S, v), label = f"{n}")

plt.xlabel("[S](M)")
plt.ylabel("Kinetic Order")
plt.legend()
```
