```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def μ_langevin(μ0, τ, α, δ):
    return μ0*np.exp(-δ*τ) + α/δ

def μ_master(μ0, τ, α, δ):
    """
    In a process where
    ø -> n, α
    n -> n - 1, δn
    we would have
    μ -> α/δ as τ -> inf
    """
    return np.sqrt(2*(α/δ)*τ + μ0**2)

def σ_langevin(D, τ, α, δ):
    """
    D here would be α2 = α + δn
    """
    return (D/(2*δ)) * (1 - np.exp(-2*δ*τ))
```

```python
τ_range = np.linspace(0, 20, 100)

plt.plot(τ_range, μ_master(1, τ_range, 0.5, 0.2), "k", label = "Master avg")
plt.plot(τ_range, μ_langevin(1, τ_range, 0.5, 0.2), "r", label = "Langevin avg")
plt.plot(τ_range, σ_langevin(μ_langevin(1, τ_range, 0.5, 0.2), τ_range, 0.5, 0.2), "g", label = "Langevin var")
plt.xlabel("τ")
plt.ylabel("n")
plt.legend()
```
