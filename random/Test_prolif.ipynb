```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 50)
α = 0.5
N0 = 5
N = N0 * np.exp(α * t)

plt.plot(t, N, "k-.", label = "Exp growth")
plt.xlabel("t")
plt.ylabel("N(t)")
plt.legend()
```

If on the other hand, we have limitations on growth, we are going to get a logistic eq.
Here, we have $n = N/N_{max}$, which is going to refer to the fraction of max population that exists at a certain time. We are then getting a logistic function like the following
$$
n = \frac{1}{1 + e^{-\tau}}
$$
where $\tau$ is time evaluated in terms of $\alpha$, as $\tau = \alpha t$.

```python
τ = α * t
n = 1/(1 + np.exp(-τ))
plt.plot(τ, n, "k-.")
plt.xlabel("τ")
plt.ylabel("n(τ)")
```
